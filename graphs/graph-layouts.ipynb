Sample Data 

```python
x = [0,1.4,4.5,3.1,9.5,3.5,7,6.2,10.43,11,12,13]
y = [0,1.3,2.3,1.5,7.8,4.3,8.1,4,9.45,11,12.5,13.6]

barx = ['Cat 1', 'Cat 2', 'Cat 3', 'Cat 4']
bary = [5,9,14,18]
```

Custom Graph Class

```python
sampleScatter = Graph(x, y, 'X', 'Y')
sampleScatter.plotScatter('dark', ',')
sampleScatter.plotScatter('light', ',')
sampleScatter.plotScatter('dark', 'x', trendline=False)

sampleBar = Graph(barx, bary, 'Qualitative', 'Quantitative')
sampleBar.plotBar('dark',horiz=True)
sampleBar.plotBar('light', customTitle='Custom Title')
```
