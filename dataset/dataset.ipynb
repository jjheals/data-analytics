```python
from statistics import mean, median, mode, variance, stdev, correlation
```

```python
x1 = [1,2,3,4]
y1 = [1,2,3,4]
```

```python
class DataSet:
    
    def __init__(self, xlist:list, ylist:list):
        self.xlist = xlist
        self.ylist = ylist
    

    # ----- =+= -----#
    # Calculate statistics for this data
    #   num of data points (lenth of x)
    #   data type of x & y [in the form of (x,y) where x and y are datatypes ???]
    #   Measures of center: mean, median, mode, range (max and min) - of y
    #   Measures of spread: variance, standard deviation - of y 
    #   correlation (if applicable - neither x or y can be strings, they must be int or float)
    def calcStats(self):
        xlen = len(self.xlist)
        ylen = len(self.ylist)
        
        if xlen != ylen:
            print('Error: The length of x and y must be equal.')
            return
        
        def checkTypes(lst:list):
            c = type(lst[0])
            for l in lst:
                if type(l) != c:
                    return False
            return True
        
        # Check if the lists are valid and consist of a single data type (each)
        if not checkTypes(self.xlist) or not checkTypes(self.ylist): 
            print('Error: The data types must be consistent within lists.')
            return
        
        # Make sure the y data is int or float so that it can be analyzed and compared
        if type(self.ylist[0]) != int and type(self.ylist[0]) != float:
            print('Error: Data list y is invalid (not int or float).')
            return
            
        
        print(f'''
Number of data points: {xlen}
x datatype: {type(self.xlist[0])}
y datatype: {type(self.ylist[0])}

NOTE - All statistics are of list y
Mean: {mean(self.ylist)}
Median: {median(self.ylist)}
Mode: {mode(self.ylist)}
Range: [{min(self.ylist)}, {max(self.ylist)}]
Variance: {variance(self.ylist)}
Standard Deviation: {stdev(self.ylist)}
              ''')
        
        if (type(self.xlist[0]) != int or type(self.xlist[0]) != float):
            print(f'Correlation of x and y: {correlation(self.xlist, self.ylist)}')
        else: print('Correlation not applicable.')

```

```python
d1 = DataSet(x1, y1)
d1.calcStats()
```
